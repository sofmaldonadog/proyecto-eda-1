```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import textblob as tb
```

```python
df = pd.read_json("News_Category_Dataset_v3.json", lines=True)
```

```python
df.head()
```

```python
#Encontrar que autores son m√°s prominentes
autores_cuenta = df["authors"].value_counts()
autores_ordenados = autores_cuenta.sort_values(ascending=False)
top_10 = autores_ordenados.head(10)
```

```python
top_10.plot(kind="bar")
plt.title("Autores m√°s comunes")
plt.show()
```

```python
df["authors"].nunique()
```

```python
#Encontrar las categor√≠as m√°s repetidas
categorias_cuenta = df["category"].value_counts()
categorias_ordenadas = categorias_cuenta.sort_values(ascending=False)
top_10 = categorias_ordenadas.head(10)
```

```python
top_10.plot(kind="bar")
plt.title("Categor√≠as m√°s comunes")
plt.show()
```
